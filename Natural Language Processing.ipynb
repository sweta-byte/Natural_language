```python
#nltk----> Natural Language Toolkit
#spacy
#gensim
#stanfordNLP

from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

documents=['Hyderabad is famous for biryani.I love to stay here',
          'I live in Hyderabad.I would recommand Hyderabad to anyone',
          'I started to like Hyderabad for its Biryani',
          'The next IPL match is in Hyderabad.I enjoy watching IPL here in Hyderabad']

cv=CountVectorizer()

dt_matrix=cv.fit_transform(documents)

df=pd.DataFrame(dt_matrix.toarray(),columns=cv.get_feature_names())
df
```

```python
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
from nltk.tag import pos_tag

sentence='Donald Trump is Visiting India on 10th December 2021.'

stopwords_en=stopwords.words('english') # stopwords of english language
tokens=word_tokenize(sentence) # word tokenizing

pos_tag(tokens) # extracting the parts of speech of every term in document
```

```python
para='tiramisu is my favourite dessert. The three layered cake with coffee powder on it would taste awesome'
sent_tokens=sent_tokenize(para)
sent_tokens
```

```python
fresh_tokens=[token for token in tokens if token.lower() not in stopwords_en]
fresh_tokens
```
